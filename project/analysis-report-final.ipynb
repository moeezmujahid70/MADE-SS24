### Limitations:

- **Linearity:** The study assumes a linear relationship between CO2 emissions and temperature thus the model does not account for most of the dynamics involved in climate change.

- **Quality of Data:** Different countries may have reported or documented data in inconsistent or using different measures, thus resulting in inconsistency in data.

- **Resolution Time:** There is no possibility of doing a short-term analysis as the dataset offers only annual data.
- **Application and Use:** Dataset sources for non-commercial use limit the data from being used for commercial application ideas.

### Future Outlook:
